```python

df['date'] = pd.to_datetime(df['date'], format='%d/%m/%Y')

df['consumption_mwh'] = df['consumption_mwh'].str.replace(',', '').astype(float)

grouped_df = df.groupby(df['date'].dt.year)['consumption_mwh'].mean()

print(grouped_df)
```

```python
plt.bar(grouped_df.index, grouped_df.values)
plt.xlabel('Yıl')
plt.ylabel('Ortalama Tüketim (mWh)')


plt.title('Yıllara Göre Ortalama Tüketim')

plt.show()
```

```python
plt.plot(grouped_df.index, grouped_df.values, marker='o')

plt.xlabel('Yıl')
plt.ylabel('Ortalama Tüketim (mWh)')

plt.title('Yıllara Göre Ortalama Tüketim')

plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X = df['date'].dt.year.values.reshape(-1, 1)
y = df['consumption_mwh'].values
model = LinearRegression()
model.fit(X, y)

ileriki_yillar = [[2024], [2025], [2026], [2027]] 
tahminler = model.predict(ileriki_yillar)

plt.plot(df['date'].dt.year, df['consumption_mwh'], marker='o', label='Mevcut Veriler')

plt.plot(ileriki_yillar, tahminler, marker='o', linestyle='dashed', label='Tahminler')

plt.xlabel('Yıl')
plt.ylabel('Ortalama Tüketim (mWh)')
plt.title('Mevcut Veriler ve Tahminler')

plt.legend()

plt.show()
```
